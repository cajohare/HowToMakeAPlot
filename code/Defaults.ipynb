# Why to use matplotlib style sheet

```python
from numpy import *
import matplotlib.pyplot as plt
from scipy.special import jn


plt.style.use('default')

fig = plt.figure()
ax = fig.add_subplot()

x = linspace(0,20,1000)

for n in range(5):
    plt.plot(x,jn(n,x),'-',label=n)

plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$j_n(x)$')

plt.savefig('../plots/DefaultPlot_Bad.pdf',transparent=False,bbox_inches='tight')
```

# With the style sheet added

```python
import matplotlib.cm as cm

plt.style.use('sty.mplstyle')


fig = plt.figure(figsize=(13,10))
ax = fig.add_subplot()

for n in range(5):
    plt.plot(x,jn(n,x),'-',label=n)

plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$j_n(x)$')
plt.xlim(left=0,right=20)


plt.savefig('../plots/DefaultPlot_Better.pdf',transparent=False,bbox_inches='tight')
```
