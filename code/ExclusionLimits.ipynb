```python
from numpy import *
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe


plt.style.use('sty.mplstyle')

def PlotConstraint(ax,filepath,lw=2.5,color='k',filled=True,fill_zorder=-1,zorder=0,y2=1,
                   facecolor='green',edgecolor='darkgreen',
                   facealpha=1,edgealpha=1,path_effects=[pe.Stroke(linewidth=4, foreground='k'), pe.Normal()],
                    text_on=True,label='',label_pos=[1e1,1e-33],fontsize=22,
                   rotation=30,text_path_effects=[pe.Stroke(linewidth=1, foreground='k'), pe.Normal()],textalpha=1,linestyle='-'):

    dat = loadtxt(filepath)
    if filled:
        plt.fill_between(dat[:,0],dat[:,1],y2=y2,color=facecolor,alpha=facealpha,zorder=fill_zorder,lw=0)
    plt.plot(dat[:,0],dat[:,1],linestyle=linestyle,color=edgecolor,lw=lw,path_effects=path_effects,alpha=edgealpha,zorder=zorder)

    if text_on:
        plt.text(label_pos[0],label_pos[1],label,fontsize=fontsize,
                 color=edgecolor,rotation=rotation,path_effects=text_path_effects,clip_on=True,alpha=textalpha,rotation_mode='anchor');
    return 


fig = plt.figure(figsize=(15,11))
ax = plt.subplot()

PlotConstraint(ax,'../data/WIMPLimits/XENONnT_non-power-constrained.txt',label='XENONnt',label_pos=[28,3e-47],rotation=13,filled=False,edgecolor='#32a852',fontsize=20)
PlotConstraint(ax,'../data/WIMPLimits/LZ_2022_non-power-constrained.txt',label='LZ',label_pos=[28,0.8e-47],rotation=22,filled=False,edgecolor='#db851d')
PlotConstraint(ax,'../data/WIMPLimits/DarkSide.txt',filled=False,edgecolor='#38c771',label='DarkSide',label_pos=[1.7,5e-42],rotation=-42)
PlotConstraint(ax,'../data/WIMPLimits/DEAP-3600.txt',filled=False,edgecolor='#0b568c',label='DEAP-3600',label_pos=[300,0.7e-44],rotation=15.5)
PlotConstraint(ax,'../data/WIMPLimits/EDELWEISS.txt',filled=False,edgecolor='#a89016',label='EDELWEISS',label_pos=[9.5,1.5e-42],rotation=-48,fontsize=15)
PlotConstraint(ax,'../data/WIMPLimits/PandaX.txt',filled=False,edgecolor='#bf041a',label='PandaX',label_pos=[120,1.9e-46],rotation=17)
PlotConstraint(ax,'../data/WIMPLimits/PICO2L.txt',filled=False,edgecolor='#8c3869',label='PICO2L',label_pos=[300,8.5e-42],rotation=17)
PlotConstraint(ax,'../data/WIMPLimits/PICO60.txt',filled=False,edgecolor='#8c3869',label='PICO60',label_pos=[350,8e-44],rotation=17.5)
PlotConstraint(ax,'../data/WIMPLimits/COSINE-100.txt',filled=False,edgecolor='#bdbd2f',label='COSINE-100',label_pos=[250,2e-42],rotation=17.5)
PlotConstraint(ax,'../data/WIMPLimits/CRESST.txt',filled=False,edgecolor='#ff5500',label='CRESST',label_pos=[2.5,6e-41],rotation=-38,fontsize=18)
PlotConstraint(ax,'../data/WIMPLimits/nuFog_Xe_SI.txt',filled=False,edgecolor='k',label='Neutrino fog (Xe)',label_pos=[0.7,2e-46],rotation=0)


ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlim([0.5,1e4])
ax.set_ylim([5e-50,1e-40])
ax.set_xlabel('Dark matter mass [GeV]')
ax.set_ylabel(r'SI nucleon cross section [cm$^{2}$]')

locmaj = matplotlib.ticker.LogLocator(base=10.0, subs=(1.0, ), numticks=50)
locmin = matplotlib.ticker.LogLocator(base=10.0, subs=arange(2, 10)*.1,numticks=100)
ax.yaxis.set_major_locator(locmaj)
ax.yaxis.set_minor_locator(locmin)
plt.savefig('../plots/ExclusionLimits_Bad.pdf',transparent=False,bbox_inches='tight')
plt.savefig('../plots/plots_png/ExclusionLimits_Bad.png',transparent=False,bbox_inches='tight')
```
