# Visual clutter

Every plot is built out of pre-attentive visual attributes automatically. Your task is to control which ones you use and to remove ones which appear by accident. Make sure every element is purposeful and supports the main message. The reason most bad plots look bad is because of sloppy use of these attributes, i.e. visual clutter. Individually these may seem like small things, but our brains are tuned to spot anything out of place. Even seemingly trivial things like asymmetries, misalignment, or overlapping elements can stack up and degrade the intelligibility of a plot. 


### Example 1: Here is an extremely simple plot to showcase the detrimental effect of having a lot of unecessary visual clutter, which often enters into plots unintentionally.

This first plot is intentonally bad. It is filled with defects, asymmetries and overlapping elements.

```python
from numpy import *
import matplotlib.pyplot as plt
from numpy import random


plt.style.use('default')

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()

# Making the data for the plot
x = array([10,20,30,40,50])
y = array([1,3,5,4,7])
yerr = array([0.2,0.4,0.6,0.6,0.7])
plt.errorbar(x,y,yerr,fmt='.',label='Data',color='k',elinewidth=1)
xvals = linspace(0,50,100)
plt.plot(xvals,0.12*xvals,color='k')

plt.xlim(left=0)
plt.ylim(bottom=0)

plt.ylabel('y',labelpad=-6,fontsize=20)
plt.xlabel('x',labelpad=-6,fontsize=20)
plt.legend(fontsize=25,loc='upper right')
ax.tick_params(length=5,width=5,colors='k',tickdir='in',top=True)
plt.text(12,2,'Fit',fontsize=20)
plt.text(30,1,'My measurement',color='b',fontsize=15)

ax.spines['bottom'].set_color('gray')
ax.spines['top'].set_color('gray') 
ax.spines['right'].set_color('gray')
ax.spines['left'].set_color('gray')
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)

plt.savefig('../plots/Clutter_Bad.pdf',transparent=False,bbox_inches='tight')
```

### Without adding or subtracting anything of any material, you can instantly improve the visual just by removing visual elements that were in the plot unintentionally.

This plot has a lot more symmetry, there are no clashing elements, and things like colors and fontsizes are used with intention to add information or to add important layering or contrast.

I have purposefully not used the style sheet in this case to show that you can improve the look just by moving things around and not changing any default settings or doinga anything fancy

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()

# plot data again
plt.errorbar(x,y,yerr,fmt='.',label='Data',color='royalblue',markersize=11)
plt.plot(xvals,0.12*xvals,'--',color='orangered')

# add elements
plt.ylabel('$y$',fontsize=23)
plt.xlabel('$x$',fontsize=23)
plt.legend(fontsize=18,loc='upper left',handletextpad=0,bbox_to_anchor=(-0.042,0.895),frameon=False)
ax.tick_params(length=5,tickdir='in',top=True,right=True,labelsize=15)
plt.text(12,1.9,'Fit',fontsize=20,rotation=35.5,color='orangered')
plt.text(0,7.2,r'My measurement',color='k',fontsize=23,fontweight='bold')

# Make axes slightly thicker than ticks
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1.5)

plt.savefig('../plots/Clutter_Better.pdf',transparent=False,bbox_inches='tight')
```
